```python
import pandas as pd

# 1. Создаем данные
data = {
    'ip_source': ['192.168.1.1', '192.168.1.5', '10.0.0.1', '192.168.1.1'],
    'packet_size': [64, 1500, 40, 1200],
    'is_anomaly': [False, False, True, False]
}

df = pd.DataFrame(data)

# 2. Выводим таблицу (в Jupyter достаточно просто написать df)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['packet_size'], marker = 'o', linestyle='--', color='red')

plt.title("Визуализация сетевого трафика")
plt.xlabel("Номер пакета")
plt.ylabel("Размер (байт)")
plt.grid(True)

plt.show
```
