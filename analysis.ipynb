```python
import pandas as pd

file_path = 'data/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv'

df_ids = pd.read_csv(file_path)

print(df_ids.columns)

df_ids.head()
```

```python
print("Распределение типов трафика:")
print(df_ids[' Label'].value_counts())

df_ids.describe()
```

```python
import numpy as np

df_ids.replace([np.inf, -np.inf], np.nan, inplace = True)

df_ids.dropna(inplace=True)

print(f"Очистка завершена. Осталось строк: {len(df_ids)}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Берем только два типа для сравнения
plt.figure(figsize=(12, 6))

# Рисуем распределение длительности потока для BENIGN и DDoS
# Ограничим до 1 000 000, чтобы график был понятным
sns.boxplot(x=' Label', y=' Flow Duration', data=df_ids)
plt.ylim(0, 1000000) 
plt.title("Сравнение длительности соединений: Normal vs DDoS")
plt.show()
```

```python
# Удаляем лишние пробелы из названий всех колонок
df_ids.columns = df_ids.columns.str.strip()

# Проверим, как теперь называются наши колонки (пробелов быть не должно)
print(df_ids.columns.tolist())
```

```python
# Теперь названия чистые, без пробелов в начале
features = ['Flow Duration', 'Total Fwd Packets', 'Total Backward Packets', 
            'Total Length of Fwd Packets', 'Fwd Packet Length Max', 'Label']

df_mini = df_ids[features].copy()

# Превращаем Label в числа: BENIGN = 0, DDoS = 1
df_mini['Label_Numeric'] = df_mini['Label'].apply(lambda x: 0 if x == 'BENIGN' else 1)

# Считаем корреляцию и рисуем тепловую карту
corr_matrix = df_mini.drop('Label', axis=1).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Корреляция признаков с атакой DDoS (Очищенная)")
plt.show()
```
